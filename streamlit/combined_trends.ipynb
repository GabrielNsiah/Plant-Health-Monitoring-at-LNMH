Notebook for creating Combined Trends over Time Graph

```python
import altair as alt
import pandas as pd
import streamlit as st
from base_script import return_merged_df
```

Import merged df

```python
merged_df = return_merged_df()
merged_df.sample(10)
```

```python
long_df = merged_df.melt(id_vars='recording_taken',
                  value_vars=['soil_moisture', 'temperature'],
                  var_name='Measurement',
                  value_name='Value')
```

```python
chart = alt.Chart(long_df).mark_line().encode(
    x='recording_taken:T',        # Time on the x-axis
    y='Value:Q',                 # Numerical values on the y-axis
    color='Measurement:N'        # Different lines for soil moisture and temperature
).properties(
    title='Soil Moisture and Temperature Over Time',
    width=800,
    height=400
)

chart.show()
```
